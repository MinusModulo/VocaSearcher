```python
import os
import json
import sqlite3
import re
from pathlib import Path
from datetime import datetime
import logging
```

```python
db_path = Path("../db/songs.db")
```

```python
conn = sqlite3.connect(db_path)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('SELECT COUNT(*) FROM songs')
count = cursor.fetchone()[0]
count # 这个数差不多是对的
```

```python
cursor.execute("SELECT COUNT(*) FROM lyrics")
count = cursor.fetchone()[0]
count
```

```python
cursor.execute("SELECT name, song_count, total_views FROM singers ORDER BY song_count DESC")
singer = cursor.fetchall()
with open("singers.txt", "w", encoding="utf-8") as f:
    for i, (name, song_count, total_views) in enumerate(singer, 1):
        f.write(f"{i}. {name} - {song_count} songs, {total_views} views\n")
    
```

```python
cursor.execute("SELECT COUNT(*) FROM word_lyrics")
count = cursor.fetchone()[0]
count
```

```python
cursor.execute("SELECT lemma, frequency FROM words ORDER BY frequency DESC")
words = cursor.fetchall()
with open("words.txt", "w", encoding="utf-8") as f:
    for lemma, frequency in words:
        f.write(f"{lemma} - {frequency}\n")
```
